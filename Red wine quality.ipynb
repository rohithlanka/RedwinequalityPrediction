```python
#importing libraries necessary for this project
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
#dropping the duplicate values
data=data.drop_duplicates()
```

```python
#displaying the first 10 rows in the dataset
data.head(10)
```

```python
data.quality.value_counts()
```

```python
#splitting the input and output variables
qua=data['quality']

features=data.drop('quality',axis=1)
#encoding the output based on quality
quality = qua.apply(lambda x: 1 if x > 5 else 0)
```

here the quality attribute  is encoded as 0,1.

```python
pd.DataFrame(quality.value_counts())
```

```python
#scatter matrix for eah pair of features
pd.scatter_matrix(features, alpha = 0.3, figsize = (14,8), diagonal = 'kde');
```

Normalise the data

```python
#normalising the data
from sklearn.preprocessing import normalize
feature=pd.DataFrame(normalize(features))
```

```python
#scatter matrix for eah pair of features
pd.scatter_matrix(feature, alpha = 0.3, figsize = (14,8), diagonal = 'kde');
```

```python
feature.head(10)
```

spliting the data into training and testing

```python
#splitting the data into training and testing 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature, quality, test_size = 0.2, random_state = 0)
```

# Implementation
## Implementing Benchmark model

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
#applying svc
clf = SVC(random_state =40)
learner =clf.fit(X_train,y_train)
predictSVC = learner.predict(X_test)
#calculating score
print(accuracy_score(y_test,predictSVC))
```

# Implementing 4 classification algorithms and checking results with accuracy score

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

#applying adaboostclassifier,logistic regression,decision tree classifier,random forest classifier
clf_A = AdaBoostClassifier(random_state = 42)
clf_B = LogisticRegression(random_state = 41)
clf_C = DecisionTreeClassifier(random_state = 41)
clf_D = RandomForestClassifier(n_estimators=200)
learner =clf_A.fit(X_train,y_train)
predict = learner.predict(X_test)
learner =clf_B.fit(X_train,y_train)
predict1 = learner.predict(X_test)
learner =clf_C.fit(X_train,y_train)
predict2 = learner.predict(X_test)
learner =clf_D.fit(X_train,y_train)
predict3 = learner.predict(X_test)
```

```python
# finding accuracy for adaboostclassifier,logistic regression,desicion tree classifier,random forest classifier
print(accuracy_score(y_test,predict))
print(accuracy_score(y_test,predict1))
print(accuracy_score(y_test,predict2))
print(accuracy_score(y_test,predict3))
```

## Tuning the model with best accuracy score by using GridSearchCV

```python
#from sklearn.model_selection import GridSearchCV
param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'criterion' :['gini', 'entropy']
}
#initialising the classifier
rfc = RandomForestClassifier(n_estimators=200)
#performing GridSearchCV for the model
grid_obj = GridSearchCV(rfc,param_grid=param_grid, cv=2)
#fitting the grid search on the classifier
grid_fit = grid_obj.fit(X_train, y_train)
#get the estimator
best_rfc = grid_fit.best_estimator_

predictions = (rfc.fit(X_train, y_train)).predict(X_test)
#making predictions
best_predictions = best_rfc.predict(X_test)
##calculating the accuracy score
print(accuracy_score(y_test,best_predictions))
```
